```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
def get_df_P0(num, group_size=36):
    # Directory where the CSV files are located
    directory = f"clean_csv/P0{num}"

    # Pattern to match CSV files
    file_pattern = "*.csv"

    # Get the file names matching the pattern for P03
    file_names = glob.glob(directory + "/" + file_pattern)


    main_dfs = []                                           #splitting up the df in case it gets too intensive (it did in the past)
    df_indx = -1                                            #index for the list above

    for indx, file_name in enumerate(file_names):
        # Read the CSV file
        df = pd.read_csv(file_name)
        df = df.drop(columns="Unnamed: 0")

        #keep only document name
        file_name=file_name.split("\\")[1]
        print(file_name)

        if indx % group_size == 0:
            df_indx += 1
            main_dfs.append(df)
            continue

        if "TraceabilityCode" in df.columns:                #check for traceability code in the columns
            merge_columns = ["TraceabilityCode", "time"]

            if "time.value" in df.columns:                  #check for time.value in the columns to avoid duplication
                merge_columns.append("time.value")
            else:
                df = df.iloc[:, :3]                         #if not time.value, drop any extra columns
            
            main_dfs[df_indx] = pd.merge(main_dfs[df_indx],df,how="outer",on=merge_columns,suffixes=("",f"_{indx}"))
            main_dfs[df_indx] = main_dfs[df_indx].set_index("TraceabilityCode")
    
    return main_dfs
```

```python
def save_df_P0(main_dfs, num):
    for i, df in enumerate(main_dfs):
        print(f"{num}_{i}")
        if i >= 1:
            df.to_csv(f"final_csv/P0{num}_{i}.csv")
        else:
            df.to_csv(f"final_csv/P0{num}.csv")
```

```python
for i in range (2,7):                       #looping through every P0_ section of files, excluding P01
    df = get_df_P0(i,36)                    #merging every p0_ dataset together into one df, separated by P0_
    
    save_df_P0(df, i)                       #saving them as csv files
```

Check for NaNs in Rows and Columns

```python
def row_nans(df):                                       #return series of rows and their percentage of nans
    rows = df.isna().sum(axis=1)/df.shape[1]
    return rows.sort_values()

def col_nans(df):                                       #return series of columns with the percentage of nans
    cols = df.isna().sum()/df.shape[0]
    return cols.sort_values()
```
